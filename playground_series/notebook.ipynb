## Data analysis generated dataset

Let's start by taking a look at the data. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("data/train.csv",index_col="id")
```

```python
df.head(5)
```

```python
X = df.drop(columns="MedHouseVal")
y = df["MedHouseVal"]
```

```python
df.info()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from tensorflow.keras import Sequential
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the model
model = tf.keras.Sequential()

# Normalization layer
model.add(tf.keras.layers.BatchNormalization(input_shape=[8]))

# Hidden layers
model.add(tf.keras.layers.Dense(units=8, activation='relu'))
model.add(tf.keras.layers.Dense(units=4, activation='relu'))
model.add(tf.keras.layers.Dense(units=4, activation='relu'))
model.add(tf.keras.layers.Dense(units=2, activation='relu'))



# Output layer
model.add(tf.keras.layers.Dense(units=1,activation='relu'))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model.fit(X_train,y_train,epochs=250, validation_split=0.2,batch_size=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
test_data= pd.read_csv("data/test.csv", index_col="id")
```

```python
# Generate test predictions
preds_test = model.predict(test_data)
output = pd.DataFrame(preds_test)
output["id"] = test_data.index
output["MedHouseVal"] = output[0] 
output.drop(columns=0, inplace=True)
```
